```python
from QuantumWalkModule.EM_OneDimensionQW import *
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
plt.rc('font', family='serif')
```

```python
steps = 100
dim = 100
coin_conf = {
    "th" : np.pi/3
}

ratios = [0.1,np.sqrt(2)/5,1/3,np.sqrt(5)]
ratio_titles = [r"$0.1$" , r"$\sqrt{2}/5$" ,r"$0.5$",r"$\sqrt{5}$"]

fig = plt.figure(figsize = (10,2))

for i, ratio in enumerate(ratios):
    field_conf = {
        "name" : "uniform",
        "ratio" : ratio
    }
    model = Electric_QW_1d(dim , coin_conf,field_conf)
    data = model.run(steps,state = "Gaussian",mu = dim // 2,sigma = dim /50,a = 1j,b = -1j)
    ax = fig.add_subplot(1,4,i + 1)
    im = ax.imshow(data, origin = "lower",cmap = "viridis",alpha = 0.95)
    plt.colorbar(im, fraction=0.05, pad=0.09)
    ax.set_xlabel("x")
    ax.set_ylabel("t")
    ax.set_title(r"$\phi / 2\pi = $" + ratio_titles[i])
plt.tight_layout()
plt.show()

fig.savefig("Images/ElectricWalks1.pdf",format= "pdf",bbox_inches = "tight")
```
